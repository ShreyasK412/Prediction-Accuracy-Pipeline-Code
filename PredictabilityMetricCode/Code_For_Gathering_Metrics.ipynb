```python
import pandas as pd
from google.colab import files

# 1. Upload the two CSV files
uploaded_files = files.upload()
# Load the FixedLSTMPredictionsTrainonEntireDataSet.csv file
data = pd.read_csv('Unbalanced_24_Predictions.csv')

# Function to compute the accuracy for each group
def compute_accuracy(group):
    correct = group[group['predicted_cluster'] == group['actual_cluster']]
    accuracy = len(correct) / len(group)
    return pd.Series({
        'accuracy': accuracy,
        'count': len(group),
        'correct_count': len(correct)
    })

# Group by seg_id and compute the accuracy metrics
results = data.groupby('seg_id').apply(compute_accuracy).reset_index()

# Save the results to ProportionResultsTrainonEntireDataSet.csv
results.to_csv('ProportionResultsUnbalanced_24.csv', index=False)
files.download('ProportionResultsUnbalanced_24.csv')
```

```python
import pandas as pd
from google.colab import files

# 1. Upload the two CSV files
#uploaded_files = files.upload()
# Load the three datasets
df_6 = pd.read_csv('ProportionResultsUnbalanced_24.csv')
df_12 = pd.read_csv('ProportionResultsUnbalanced_12.csv')
df_24 = pd.read_csv('ProportionResultsUnbalanced_6.csv')

# Merge the dataframes on 'seg_id'
merged_df = df_6.merge(df_12, on='seg_id', how='inner', suffixes=('_6', '_12'))
merged_df = merged_df.merge(df_24, on='seg_id', how='inner')

# Ensure only the accuracy columns are selected and compute the average
merged_df['average_accuracy'] = (merged_df['accuracy_6'] + merged_df['accuracy_12'] + merged_df['accuracy']) / 3

# Select only the 'seg_id' and 'average_accuracy' columns for the final output
final_df = merged_df[['seg_id', 'average_accuracy']]

# Save the resulting dataframe to a CSV
final_df.to_csv('UnbalancedAverageAccuracies.csv', index=False)
```
