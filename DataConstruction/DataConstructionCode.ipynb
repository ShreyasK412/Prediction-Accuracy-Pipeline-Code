```python
import pandas as pd
from google.colab import files

uploaded = files.upload()

# Assuming you've uploaded a single file named 'your_file.csv'
filename = list(uploaded.keys())[0]
df = pd.read_csv(filename)

# Display the first few rows of the dataframe
print(df.head())
```

```python
df=df.drop(['coef_quadratic', 'coef_linear','coef_intercept'], axis=1)



df.head()
```

```python
df=df.drop('n_qshapes', axis=1)
df.head()
```

```python
df=df.drop(['exp_run_id','duration_ms'], axis=1)
```

```python
import pandas as pd

# Your sample dataframe


# Initialize lists to store the sequences and targets
sequences = []
targets = []

# Identify where the sequence resets
reset_indices = df[df['time_bin_id'] == 1].index.tolist()

# Append a sentinel value to handle the last sequence
reset_indices.append(len(df))

for i in range(len(reset_indices) - 1):
    start = reset_indices[i]
    end = reset_indices[i + 1]
    seq = df['cluster'][start:end].tolist()

    for j in range(len(seq) - 10):
        sequences.append(seq[j:j+10])
        targets.append(seq[j+10])

# Convert to DataFrame
result_df = pd.DataFrame({'sequence': sequences, 'target': targets})

print(result_df)
```

```python
# Assuming df is your existing dataframe
# Initialize lists to store the sequences, targets, and the additional features
sequences = []
targets = []
seg_ids = []
speaker_ids = []
datasets = []
speech_registers = []

# Identify where the sequence resets
reset_indices = df[df['time_bin_id'] == 1].index.tolist()

# Append a sentinel value to handle the last sequence
reset_indices.append(len(df))

for i in range(len(reset_indices) - 1):
    start = reset_indices[i]
    end = reset_indices[i + 1]
    seq = df['cluster'][start:end].tolist()

    for j in range(len(seq) - 10):
        sequences.append(seq[j:j+10])
        targets.append(seq[j+10])

        # Extracting the additional features for each sequence
        seg_ids.append(df['seg_id'].iloc[start])
        speaker_ids.append(df['speaker_id'].iloc[start])
        datasets.append(df['dataset'].iloc[start])
        speech_registers.append(df['speech_register'].iloc[start])

# Convert to DataFrame
result_df = pd.DataFrame({
    'sequence': sequences,
    'target': targets,
    'seg_id': seg_ids,
    'speaker_id': speaker_ids,
    'dataset': datasets,
    'speech_register': speech_registers
})

print(result_df)
```
