```python
import pandas as pd
from google.colab import files

uploaded = files.upload()

# Assuming you've uploaded a single file named 'your_file.csv'
filename = list(uploaded.keys())[0]
df = pd.read_csv(filename)

# Display the first few rows of the dataframe
print(df.head())
```

```python
df=df.drop(['coef_quadratic', 'coef_linear','coef_intercept'], axis=1)
```

```python
df=df.drop('n_qshapes', axis=1)
df.head()
```

```python
df=df.drop(['exp_run_id','duration_ms'], axis=1)
```

```python
import pandas as pd

# Your sample dataframe


# Initialize lists to store the sequences and targets
sequences = []
targets = []

# Identify where the sequence resets
reset_indices = df[df['time_bin_id'] == 1].index.tolist()

# Append a sentinel value to handle the last sequence
reset_indices.append(len(df))

for i in range(len(reset_indices) - 1):
    start = reset_indices[i]
    end = reset_indices[i + 1]
    seq = df['cluster'][start:end].tolist()

    for j in range(len(seq) - 10):
        sequences.append(seq[j:j+10])
        targets.append(seq[j+10])

# Convert to DataFrame
result_df = pd.DataFrame({'sequence': sequences, 'target': targets})

print(result_df)
```

```python
# Assuming df is your existing dataframe
# Initialize lists to store the sequences, targets, and the additional features
sequences = []
targets = []
seg_ids = []
speaker_ids = []
datasets = []
speech_registers = []

# Identify where the sequence resets
reset_indices = df[df['time_bin_id'] == 1].index.tolist()

# Append a sentinel value to handle the last sequence
reset_indices.append(len(df))

for i in range(len(reset_indices) - 1):
    start = reset_indices[i]
    end = reset_indices[i + 1]
    seq = df['cluster'][start:end].tolist()

    for j in range(len(seq) - 10):
        sequences.append(seq[j:j+10])
        targets.append(seq[j+10])

        # Extracting the additional features for each sequence
        seg_ids.append(df['seg_id'].iloc[start])
        speaker_ids.append(df['speaker_id'].iloc[start])
        datasets.append(df['dataset'].iloc[start])
        speech_registers.append(df['speech_register'].iloc[start])

# Convert to DataFrame
result_df = pd.DataFrame({
    'sequence': sequences,
    'target': targets,
    'seg_id': seg_ids,
    'speaker_id': speaker_ids,
    'dataset': datasets,
    'speech_register': speech_registers
})

print(result_df)
```

```python
import pandas as pd
import pandas as pd
from google.colab import files

uploaded = files.upload()

# Load the dataset
df = pd.read_csv('TrainData6Clusters.csv')

# Count the number of instances in each class
ids_count = df[df['speech_register'] == 'IDS'].shape[0]
ads_count = df[df['speech_register'] == 'ADS'].shape[0]

# Determine the number of samples to match (the smaller count)
n_samples = min(ids_count, ads_count)

# Randomly sample from the majority class
if ids_count > ads_count:
    df_ids = df[df['speech_register'] == 'IDS'].sample(n_samples, random_state=42)
    df_ads = df[df['speech_register'] == 'ADS']
else:
    df_ids = df[df['speech_register'] == 'IDS']
    df_ads = df[df['speech_register'] == 'ADS'].sample(n_samples, random_state=42)

# Concatenate the balanced data
balanced_df = pd.concat([df_ids, df_ads])

# Shuffle the dataset if needed
balanced_df = balanced_df.sample(frac=1, random_state=42).reset_index(drop=True)

# Save the balanced dataset to a new CSV
balanced_df.to_csv('BalancedTrainData24Clusters.csv', index=False)
files.download('BalancedTrainData24Clusters.csv')
```

```python
import pandas as pd
from google.colab import files
# Load the dataset
uploaded = files.upload()
df = pd.read_csv('TrainData24ClustersWithGender.csv')

# Count the number of instances of MALE and FEMALE
male_count = df[df['adu_gender_m'] == 'MALE'].shape[0]
female_count = df[df['adu_gender_m'] == 'FEMALE'].shape[0]

# Determine the number of samples to match (the smaller count)
n_samples = min(male_count, female_count)

# Randomly sample from the majority class
if male_count > female_count:
    df_male = df[df['adu_gender_m'] == 'MALE'].sample(n_samples, random_state=42)
    df_female = df[df['adu_gender_m'] == 'FEMALE']
else:
    df_male = df[df['adu_gender_m'] == 'MALE']
    df_female = df[df['adu_gender_m'] == 'FEMALE'].sample(n_samples, random_state=42)

# Concatenate the balanced data
balanced_df = pd.concat([df_male, df_female])

# Shuffle the dataset
balanced_df = balanced_df.sample(frac=1, random_state=42).reset_index(drop=True)

# Save the balanced dataset to a new CSV
balanced_df.drop(columns=['adu_gender_m'], inplace=True)
balanced_df.to_csv('Balanced6ClustersWithGender.csv', index=False)
files.download('Balanced6ClustersWithGender.csv')
```
